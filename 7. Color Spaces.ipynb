# Color Spaces

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Photos/park.jpg', 1)
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

### BGR To HSV 

```python
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
img_hsv_rgb = cv.cvtColor(img_hsv, cv.COLOR_BGR2RGB)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(img_hsv_rgb)
axs[1].set_title('HSV Image')

plt.show()

cv.imshow('HSV Image', img_hsv)
cv.waitKey(2000)
cv.destroyAllWindows()
```

### BGR To LAB

```python
img_lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
img_lab_rgb = cv.cvtColor(img_lab, cv.COLOR_BGR2RGB)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(img_lab_rgb)
axs[1].set_title('LAB Image')

plt.show()

cv.imshow('LAB Image', img_lab)
cv.waitKey(2000)
cv.destroyAllWindows()
```
