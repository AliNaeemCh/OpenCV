# Face Detection Haar Cascade

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Photos/me.jpeg', 1)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_gray, vmin=0, vmax=255, cmap='gray')
plt.show()
```

```python
haar_cascade = cv.CascadeClassifier('misc/haar_face.xml')
face_rect = haar_cascade.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=50)
# minNeighbours parameter specifies the minimum number of neighboring rectangles that a candidate rectangle should have to retain it.
# More neighbours increases precision. Less neighbours increases recall
# This function returns a list of rectangles corresponding to the number of faces found in the image.
```

```python
img_marked = img_rgb.copy()
for x, y, w, h in face_rect:
    cv.rectangle(img_marked, (x, y), (x+w, y+h), (0, 255, 0), thickness=10)
plt.imshow(img_marked)
plt.show()
```
