# Basic Functions

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Photos/park.jpg', 1)
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

### BGR to Grayscale

```python
image_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

axs[1].imshow(image_gray, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Grayscale Image')

plt.show()
cv.imshow('Park', img)
cv.imshow('Park Gray', image_gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Blurring

```python
kernel_size = (7,11)     # More the kernel size, more is the blurring. Values kept odd numbered. 
img_blurred = cv.GaussianBlur(img, kernel_size, cv.BORDER_DEFAULT)
image_blurred_rgb = cv.cvtColor(img_blurred, cv.COLOR_BGR2RGB)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

axs[1].imshow(image_blurred_rgb, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Blurred Image')

plt.show()
cv.imshow('Park', img)
cv.imshow('Park Blurred', img_blurred)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Edge Detection

```python
lower_threshold = 250
upper_threshold = 500
# optimal lower upper threshold ratio is 1:2 or 1:3
img_edges = cv.Canny(img, lower_threshold, upper_threshold)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')

axs[1].imshow(img_edges, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Image Edges')

plt.show()
cv.imshow('Park', img)
cv.imshow('Park Edges', img_edges)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Dilation

```python
img_dilated = cv.dilate(img_edges, kernel=(7,7), iterations=3)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img_edges, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Image Edges')

axs[1].imshow(img_dilated, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Dilated Image Edges')

plt.show()
cv.imshow('Park Edges', img_edges)
cv.imshow('Dilated Park Edges', img_dilated)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Erosion

```python
img_eroded = cv.erode(img_dilated, kernel=(7,7), iterations=3)
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img_dilated, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Dilated Image Edges')

axs[1].imshow(img_eroded, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Eroded Image Edges')

plt.show()
cv.imshow('Dilated Park Edges', img_dilated)
cv.imshow('Eroded Park Edges', img_eroded)
cv.waitKey(0)
cv.destroyAllWindows()
```
