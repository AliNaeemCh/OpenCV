# Reading & Writing Images

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

### Reading as a color image

```python
img = cv.imread('Photos/cat.jpg', 1) # flag = 1 for color image.
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # Convert from BGR to RGB
print('Image shape:', img.shape)
plt.imshow(image_rgb)
plt.show()
cv.imshow('cat', img)   #First parameter is the name of the window. OpenCV reads color images as BGR
cv.waitKey(2000) # Pauses the execution for 2000ms or until any key is pressed.          
# The parameter inside waitKey is the duration in ms for which the execution will be paused unless a key is pressed.
# If any key is pressed before the specified duration, the execution resumes. Otherwise, the resumption takes place after specified time passes.
# 0 means infinite duration. So, the execution will remain paused until any key is pressed.
cv.destroyAllWindows()  # This closes all the open windows.
```

### Reading as a gray scale image

```python
img2 = cv.imread('Photos/cat.jpg', 0) # flag = 0 for gray-scale image.
print('Image shape:', img2.shape)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()
cv.imshow('cat gray', img2)
cv.waitKey(2000)
cv.destroyAllWindows()
```

### Writing Images

```python
cv.imwrite('Photos/cat_gray.jpg', img2)
```
