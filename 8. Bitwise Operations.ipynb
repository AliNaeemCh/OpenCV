# Bitwise Operations

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
blank = np.zeros((400, 400), dtype=np.uint8)
rectangle = cv.rectangle(blank.copy(), (30, 30), (370, 370), 255, -1)
circle = cv.circle(blank.copy(), (200, 200), 200, 255, -1)
```

### Bit-wise AND

```python
bitwise_AND = cv.bitwise_and(rectangle, circle)
# Create a figure
fig = plt.figure(figsize=(10, 8))

# Create a GridSpec with 2 rows and 2 columns
gs = gridspec.GridSpec(2, 2, height_ratios=[1, 1])

# Add subplots to the GridSpec
ax1 = fig.add_subplot(gs[0, 0])  # Upper row, left column
ax2 = fig.add_subplot(gs[0, 1])  # Upper row, right column
ax3 = fig.add_subplot(gs[1, :])  # Lower row, spanning both columns

# Plot something on each subplot
ax1.imshow(rectangle, cmap='gray', vmin=0, vmax=255)
ax1.set_title('Rectange')

ax2.imshow(circle, cmap='gray', vmin=0, vmax=255)
ax2.set_title('Circle')

ax3.imshow(bitwise_AND, cmap='gray', vmin=0, vmax=255)
ax3.set_title('Bit-wise AND')

# Adjust layout to prevent overlap
plt.tight_layout()

# Adjust the position of the lower subplot to center it
pos1 = ax3.get_position()  # Get the original position
pos2 = [pos1.x0 + (pos1.width / 2) - (ax1.get_position().width / 2), pos1.y0, ax1.get_position().width, pos1.height] 
ax3.set_position(pos2)  # Set a new position

# Display the plot
plt.show()
```

### Bit-wise OR

```python
bitwise_OR = cv.bitwise_or(rectangle, circle)
# Create a figure
fig = plt.figure(figsize=(10, 8))

# Create a GridSpec with 2 rows and 2 columns
gs = gridspec.GridSpec(2, 2, height_ratios=[1, 1])

# Add subplots to the GridSpec
ax1 = fig.add_subplot(gs[0, 0])  # Upper row, left column
ax2 = fig.add_subplot(gs[0, 1])  # Upper row, right column
ax3 = fig.add_subplot(gs[1, :])  # Lower row, spanning both columns

# Plot something on each subplot
ax1.imshow(rectangle, cmap='gray', vmin=0, vmax=255)
ax1.set_title('Rectange')

ax2.imshow(circle, cmap='gray', vmin=0, vmax=255)
ax2.set_title('Circle')

ax3.imshow(bitwise_OR, cmap='gray', vmin=0, vmax=255)
ax3.set_title('Bit-wise OR')

# Adjust layout to prevent overlap
plt.tight_layout()

# Adjust the position of the lower subplot to center it
pos1 = ax3.get_position()  # Get the original position
pos2 = [pos1.x0 + (pos1.width / 2) - (ax1.get_position().width / 2), pos1.y0, ax1.get_position().width, pos1.height] 
ax3.set_position(pos2)  # Set a new position

# Display the plot
plt.show()
```
