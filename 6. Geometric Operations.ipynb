# Geometric Operations

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('Photos/park.jpg', 1)
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

### Translation

```python
def translate_img(img, x, y):
    trans_mat = np.array([[1,0,x], [0,1,y]], dtype=np.float32)
    height = img.shape[0]
    width = img.shape[1]
    dims = (width, height)
    return cv.warpAffine(img, trans_mat, dims)
# x = +ve -> Right
# x = -ve -> Left
# y = +ve -> Down
# y = -ve -> Up
```

```python

img_translated_rgb = translate_img(image_rgb, 200, -100)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(img_translated_rgb)
axs[1].set_title('Translated Image')

plt.show()
```

### Rotation

```python
def rotate_img(img, angle):
    height = img.shape[0]
    width = img.shape[1]
    rot_point = (width//2, height//2)   # Rotate aroud center
    rot_mat = cv.getRotationMatrix2D(center=rot_point, angle=angle, scale=1)
    dims = (width, height)
    return cv.warpAffine(img, rot_mat, dims)
# angle = +ve -> CCW
# angle = -ve -> CW
```

```python

img_rotated_rgb = rotate_img(image_rgb, 45)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(img_rotated_rgb)
axs[1].set_title('Rotated Image')

plt.show()
```

### Flipping

```python

img_flipped_rgb = cv.flip(image_rgb, flipCode=0)

# flipCode = 0 -> Vertical Flip
# flipCode = 1 -> Horiontal Flip
# flipCode = -1 -> Both Vertical and Horizontal Flip

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[1].imshow(img_flipped_rgb)
axs[1].set_title('Flipped Image')

plt.show()
```
