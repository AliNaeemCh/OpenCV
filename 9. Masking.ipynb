# Masking

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('Photos/cats.jpg')
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # Convert from BGR to RGB
plt.imshow(image_rgb)
plt.show()
```

```python
height, width = img.shape[:2]
blank = np.zeros((height, width), dtype=np.uint8)
rectangle = cv.rectangle(blank.copy(), (300, 170), (380, 270), 255, -1)
circle = cv.circle(blank.copy(), (350, 200), 50, 255, -1)
```

### Circular Mask

```python
img_circular = cv.bitwise_and(image_rgb, image_rgb, mask=circle)
plt.imshow(img_circular)
plt.show()
```

### Rectangular Mask

```python
img_rectangular = cv.bitwise_and(image_rgb, image_rgb, mask=rectangle)
plt.imshow(img_rectangular)
plt.show()
```
