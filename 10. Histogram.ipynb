# Histogram

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('Photos/cats.jpg')
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)  # Convert from BGR to RGB
plt.imshow(image_rgb)
plt.show()
```

```python
image_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Histogram of Gray-scale Image

```python
hist_gray = cv.calcHist([image_gray], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist_gray)
plt.xlabel('Intensity')
plt.ylabel('No. of Pixels')
plt.title('Gray-scale Image Histogram')
plt.show()
```

### Histogram of Gray-scale Masked Image

It will consider only the masked portion of the image to calculate the histogram.

```python
height, width = img.shape[:2]
blank = np.zeros((height, width), dtype=np.uint8)
circle = cv.circle(blank.copy(), (350, 200), 50, 255, -1)
hist_gray = cv.calcHist([image_gray], channels=[0], mask=circle, histSize=[256], ranges=[0, 256])
plt.plot(hist_gray)
plt.xlabel('Intensity')
plt.ylabel('No. of Pixels')
plt.title('Gray-scale Masked Image Histogram')
plt.show()
```

### Histogram of Colored Image

```python
hist_r = cv.calcHist([image_rgb], [0], None, [256], [0, 256])
hist_g = cv.calcHist([image_rgb], [1], None, [256], [0, 256])
hist_b = cv.calcHist([image_rgb], [2], None, [256], [0, 256])
plt.plot(hist_r, color='r')
plt.plot(hist_g, color='g')
plt.plot(hist_b, color='b')
plt.xlabel('Intensity')
plt.ylabel('No. of Pixels')
plt.title('Colored Image Histogram')
plt.show()
```
