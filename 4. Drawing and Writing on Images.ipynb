# Drawing and Writing on Images

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
height = 500
width = 500
img = np.zeros((height, width, 3), dtype=np.uint8)  #RGB
image_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

### Drawing a rectangle outline

```python
img2 = img.copy()
# points are given in (x, y) (horizontal, vertical)
cv.rectangle(img2, pt1=(width//4, height//4), pt2=((3*width)//4, (3*height)//4), color=(0, 255, 0), thickness=2)  # color = (B, G, R)
img2_rgb = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
plt.imshow(img2_rgb)
plt.show()
```

### Drawing a filled rectangle

```python
# For a filled shape, thickness = -1
img3 = img.copy()
cv.rectangle(img3, pt1=(width//4, height//4), pt2=((3*width)//4, (3*height)//4), color=(0, 255, 0), thickness=-1) 
img3_rgb = cv.cvtColor(img3, cv.COLOR_BGR2RGB)
plt.imshow(img3)
plt.show()
```

### Drawing a circle

```python
img4 = img.copy()
cv.circle(img4, center=(width//2, height//2), radius=(width+height)//8, color=(0, 0, 255), thickness=-1) 
img4_rgb = cv.cvtColor(img4, cv.COLOR_BGR2RGB)
plt.imshow(img4_rgb)
plt.show()
```

### Drawing a line

```python
img5 = img.copy()
cv.line(img5, pt1=(0, height-1), pt2=(width//2-1, 0), color=(255, 0, 0), thickness=2) 
cv.line(img5, pt1=(width//2-1,0), pt2=(width-1, height-1), color=(255, 0, 0), thickness=2) 
cv.line(img5, pt1=(0, height//2-1), pt2=(width-1, height//2-1), color=(255, 0, 0), thickness=2) 
img5_rgb = cv.cvtColor(img5, cv.COLOR_BGR2RGB)
plt.imshow(img5_rgb)
plt.show()
```

### Writing text

```python
img6 = img.copy()
text = "Hello, OpenCV!"
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_thickness = 2
color = (255, 255, 255)  # White color

# Get the width, height of the text box, and baseline
(text_width, text_height), baseline = cv.getTextSize(text, font, font_scale, font_thickness)

# Calculate the top center position
x = (width - text_width) // 2
y = text_height + baseline  # Adjusting y to accommodate baseline

# Put the text on the image, with y adjusted for baseline
cv.putText(img6, text, (x, y), font, font_scale, color, font_thickness)
img6_rgb = cv.cvtColor(img6, cv.COLOR_BGR2RGB)
plt.imshow(img6_rgb)
plt.show()
```
