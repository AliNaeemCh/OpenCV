# Thresholding

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('Photos/cats.jpg')
image_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Manual Thresholding

```python
threshold, img_thresh = cv.threshold(image_gray, thresh=150, maxval=255, type=cv.THRESH_BINARY)
plt.imshow(img_thresh, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Adaptive Thresholding

```python
img_ad_thresh = cv.adaptiveThreshold(image_gray, maxValue=255, adaptiveMethod=cv.ADAPTIVE_THRESH_MEAN_C, thresholdType=cv.THRESH_BINARY, blockSize=15, C=0)
plt.imshow(img_ad_thresh, cmap='gray', vmin=0, vmax=255)
plt.show()
```
